```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

img_path = "../data/SSS25.png"
image = cv2.imread(img_path, cv2.IMREAD_COLOR)

if image is None:
    raise FileNotFoundError(f"Cannot load image: {img_path}")

print(image.shape)

# Split channels (BGR)
b, g, r = cv2.split(image)

# Plot all channels
plt.figure(figsize=(10, 4))

plt.subplot(1, 3, 1)
plt.title("Red Channel")
plt.imshow(r, cmap="gray")
plt.axis("off")

plt.subplot(1, 3, 2)
plt.title("Green Channel")
plt.imshow(g, cmap="gray")
plt.axis("off")

plt.subplot(1, 3, 3)
plt.title("Blue Channel")
plt.imshow(b, cmap="gray")
plt.axis("off")

plt.tight_layout()
plt.show()

print("Red max/min:", np.max(r), np.min(r))
```

```python
import random

num_pixels = 7000

x = [random.randint(0, 600) for _ in range(num_pixels)]
y = [random.randint(0, 900) for _ in range(num_pixels)]

lst_diff = []
for i in range(num_pixels):
    lst_diff.append(max(r[x[i], y[i]], g[x[i], y[i]], b[x[i], y[i]]) - min(r[x[i], y[i]], g[x[i], y[i]], b[x[i], y[i]] ))
    # print(f"Pixel {i}: R={r[x[i], y[i]]}, G={g[x[i], y[i]]}, B={b[x[i], y[i]]} DIFF={max(r[x[i], y[i]], g[x[i], y[i]], b[x[i], y[i]]) - min(r[x[i], y[i]], g[x[i], y[i]], b[x[i], y[i]] )}")

print("Max diff:", max(lst_diff))
print("Mean diff:", sum(lst_diff)/len(lst_diff))
plt.plot(lst_diff)
```

```python

rgb_image = np.stack([r, g, b], axis=2)

# Plot the image
plt.figure(figsize=(6,6))
plt.imshow(rgb_image)
plt.title("Image shown by plot (from r, g, b)")
plt.axis("off")
plt.show()
```
